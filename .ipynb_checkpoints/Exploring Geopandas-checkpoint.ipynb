```python
import geopandas  as gpd
import matplotlib.pyplot as plt
```

```python
datasource = gpd.datasets.get_path('naturalearth_lowres')
gdf = gpd.read_file(datasource)
```

```python
type(gdf)
```

```python
gdf.plot()
plt.show()
```
